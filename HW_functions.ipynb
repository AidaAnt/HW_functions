```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def get_name():
    user_input2 = input('Введите номер документа ')
    for i in documents:
        if i['number'] == user_input2:
            return i['name']
            break
    return 'Документ не найден в базе'

def get_shelf():
    user_input2 = input('Введите номер документа ')
    for key, val in directories.items():
        if user_input2 in val:
            return key
            break
    return 'Документ не найден в базе'

def print_notes(documents,directories):
    for i in documents:
        for key, val in directories.items():
            if i['number'] in val:
                print("No:", i['number'], ", тип:", i['type'], ", владелец:", i['name'], ", полка хранения:", key)

def add_shelf():
    user_input2 = input('Введите номер полки: ')
    if user_input2 in directories:
        return "Такая полка уже существует. Текущий перечень полок:", list(directories.keys())
    else:
        directories[user_input2] = " "
        return "Полка добавлена. Текущий перечень полок:", list(directories.keys())

def del_shelf(directories):
    user_input2 = input('Введите номер полки: ')
    if user_input2 in directories:
        if directories.get(user_input2) == []:
            del directories[user_input2]
            return 'Полка удалена. Текущий перечень полок:', list(directories.keys())
        elif directories.get(user_input2) != []:
            return 'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:', list(directories.keys())
    else:
        return 'Такой полки не существует. Текущий перечень полок:', list(directories.keys())

while True:
    user_input = input('Введите команду ')
    if user_input == 'p':
        print(get_name())

    elif user_input == 's':
        print(get_shelf())

    elif user_input == 'l':
        print_notes(documents, directories)

    elif user_input == 'ads':
        print(add_shelf())

    elif user_input == 'ds':
        print(del_shelf(directories))

    elif user_input == 'q':
        break
```
